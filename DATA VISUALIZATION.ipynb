#BAR PLOT EXAMPLE

```python
categories=["A","B","C","D","E"]
values=[10,20,30,40,50]

#COLOR =["RED","GREEN","YELLOW"]

plt.bar(categories,values,color=["red","green","yellow","pink","blUE"],hatch="*")
plt.title("BAR PLOT")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories=["A","B","C"]
values=[10,20,30]

plt.bar(categories,values,color=["red","green","blue"],hatch="*",label="bar")
plt.title("bar plot")
plt.xlabel("categories")
plt.ylabel("values")
plt.legend()
plt.show()
```

#SCATTER PLOT

```python
study_hours=[1,2,3,4,5]
test_scores=[60,70,80,75,90]

#scatter plot

plt.scatter(study_hours,test_scores,color="red",marker="*")
plt.title("study_hours vs test_score")
plt.xlabel("study hours")
plt.ylabel("test scores")
plt.grid(True)
plt.show()
```

```python
study_hours=[3,5,6,9]
test_scores=[40,60,70,90]

plt.scatter(study_hours,test_scores,color=["red","blue","black","yellow"],marker="*")
plt.title("study_hours vs test_hours")
plt.xlabel("study hours")
plt.ylabel("test scores")
plt.grid(True)
plt.show()
```

#SUB PLOT

```python
fig, axs=plt.subplots(1,2,figsize=(10,5))

axs[0].bar(["A","B","C"],[3,8,9],hatch="/")
axs[0].set_title("Bar Plot")

axs[1].plot([1,2,3],[6,7,8])
axs[1].set_title("Line Plot")

plt.tight_layout()
plt.show()
```

```python
fig, axs=plt.subplots(1,2,figsize=(10,6))
axs[0].bar(["A","B","C"],[3,8,9],color=["red","blue","black"],hatch="*")
axs[0].set_title("bar plot")

axs[1].plot([1,2,3],[4,5,6],color="red",marker="*")
axs[1].set_title("line plot")
plt.grid(True)

plt.tight_layout()
plt.show()
```

PIE CHART

```python
# pie chart
labels=["Apples","Bananas","Cherries","dates"]
sizes=[25,35,30,10]
plt.figure(figsize=(6,6))
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90,shadow=True)
plt.title("Fruit Distribution")
plt.show()
```

```python
labels=["Litchi","Mango","Guava","grapes"]
sizes=[20,30,40,70]
plt.figure(figsize=(6,6))
plt.pie(sizes,labels=labels,autopct="%1.1ff%%",startangle=90,shadow=True)
plt.title("FRUIT_DISTRIBUTION")
plt.show()
```

```python
#FIRST SUBPLOT

plt.figurre(figsize=(14,10))
plt.subplot(2,2,1)
plt.plot([1,2,3],[4,5,6])
plt.tiltle("LINE PLOT")

#SECOND PLOT
plt.subplot(2,2,2)
plt.bar([1,2,3],[4,5,6])
plt.title("BAR CHART")

#third subplot
plt.subplot(2,2,3)
labels=["Apples","Banans","Cherries","Dates"]
sizes=[25,35,30,10]
#plt,figure(figsize(6,6))
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90,shadow=True)
plt.title("FRUIT DISTRIBUTION")


#fourth SUBPLOT
plt.subplot(2,2,4)
#DATA
study_hours=[1,2,3,4,5]
test_scores=[60,70,80,75,85]


#create scatterplot
plt.scatter(study_hours,test_scores,color="red",marker="*")

#add label and title
plt.title("Study Hours vs Test Scores")
plt.xlabel("Study Hours")
plt.ylabel("Test Scores")

#SHOW THE PLOT
plt.grid(True)
plt.tight_layout()    # prevent over lapping title
plt.show()

```
